```sql
SELECT ol.OrderID,
CONVERT(varchar, o.OrderDate, 107) OrderDate,
c.LastName + ', ' + SUBSTRING(c.FirstName, 1, 1) CustomerName,
ol.ItemID,
i.Description ItemDescription,
ol.QtyOrdered QtyOrdered,
ISNULL(SUM(sl.QtyShipped),0) TotalQtyShipped,
(ol.QtyOrdered - ISNULL(SUM(sl.QtyShipped),0)) LeftToShip,
CASE 
WHEN (ol.QtyOrdered - ISNULL(SUM(sl.QtyShipped),0)) = 0
THEN 'Correctly Shipped'
WHEN (ol.QtyOrdered - ISNULL(SUM(sl.QtyShipped),0)) < 0
THEN 'Over Shipped'
WHEN (ol.QtyOrdered - ISNULL(SUM(sl.QtyShipped),0)) > 0
THEN 'Partially Shipped'
WHEN ISNULL(SUM(sl.QtyShipped),0) = 0
THEN 'Not Shipped'
END ShippingStatus
FROM TblOrderLine ol
FULL OUTER JOIN TblShipLine sl
ON sl.OrderID = ol.OrderID AND sl.ItemID = ol.ItemID
LEFT OUTER JOIN TblOrder o 
ON o.OrderID = ol.OrderID
LEFT OUTER JOIN TblItem i 
ON ol.ItemID = i.ItemID
LEFT OUTER JOIN TblCustomer c 
ON c.customerID = o.CustomerID
GROUP BY ol.OrderID, ol.ItemID, ol.Price, ol.QtyOrdered, LastName, FirstName, OrderDate, Description
ORDER BY ol.OrderID, ol.ItemID
```

```sql
CREATE VIEW vMostRecentLC AS
SELECT DISTINCT ItemID,
CONVERT(VARCHAR, MAX(LastCostDate), 107) LastCostDate,
(SELECT TOP 1 CONVERT(DECIMAL(8,2), LastCost) FROM TblItemCostHistory WHERE TblItemCostHistory.ItemID = ich.ItemID ORDER BY LastCostDate DESC) AS MostCurrentCost
FROM TblItemCostHistory ich
GROUP BY ItemID
```

```sql
CREATE VIEW vOrderLineSum AS
SELECT ItemID,
SUM(ol.QtyOrdered) QtyOrdered,
Count(ol.QtyOrdered) CountOfOrders,
MAX(Price) MostExpensivePrice,
MIN(Price) LeastExpensivePrice,
AVG(Price) AveragePrice
FROM TblOrderLine ol
GROUP BY ItemID
```

```sql
SELECT i.ItemID,
I.Description ItemDescription,
ISNULL(ols.QtyOrdered,0) TotalQtyOnOrder,
ISNULL(ols.CountOfOrders,0) CountOfOrders,
ISNULL((ols.MostExpensivePrice),0) MostExpensiveCost,
ISNULL(ols.AveragePrice,0) AveragePrice,
ISNULL(MAX(ich.lastcost),0) MostExpensiveCost,
ISNULL(MIN(ich.LastCost),0) LeastExpensiveCost,
ISNULL(AVG(ich.LastCost),0) AverageCost,
CASE 
WHEN lc.LastCostDate IS NULL
THEN 'No Previous Purchase'
ELSE CAST(lc.LastCostDate AS varchar)
END LastCostDate,
ISNULL(lc.MostCurrentCost,0) MostCurrentCost
FROM TblItem i
LEFT OUTER JOIN vOrderLineSum ols
ON ols.ItemID = i.ItemID
LEFT OUTER JOIN vMostRecentLC lc
ON lc.ItemID = i.ItemID
LEFT OUTER JOIN TblItemCostHistory ich
ON ich.itemid = i.itemid
GROUP BY i.ItemID, I.description, lc.lastcostdate, lc.MostCurrentCost, ols.QtyOrdered, ols.CountOfOrders, ols.MostExpensivePrice, ols.AveragePrice
```

```sql
WITH ItemQuantities AS 
(SELECT itemID,
SUM(CASE WHEN QtyOnHand > 0 THEN QtyOnHand ELSE 0 END) TotalQuantityOnHand
FROM TblItemLocation
GROUP BY ItemID)

SELECT OpenOrderItems.ItemID,
    i.DESCRIPTION ItemDescription,
    SUM(CASE WHEN OpenOrderItems.OpenOrderQty > 0 THEN OpenOrderItems.OpenOrderQty ELSE 0 END) AS OpenOrderQty,
    ISNULL(SUM(ItemQuantities.TotalQuantityOnHand),0) AS TotalAvailableInventory,
    ISNULL(SUM(OpenOrderQty - ItemQuantities.TotalQuantityOnHand),0) AS QuantityShort
FROM (SELECT ol.itemID,
    (CASE WHEN ol.QtyOrdered > ISNULL(sl.QtyShipped,0) THEN ol.QtyOrdered - ISNULL(sl.QtyShipped,0)
    ELSE 0 END) AS OpenOrderQty
    FROM TblOrderLine ol
    LEFT JOIN TblShipLine sl
    ON sl.OrderID = ol.OrderID AND sl.ItemID = ol.ItemID) AS OpenOrderItems
LEFT JOIN ItemQuantities
ON OpenOrderItems.ItemID = ItemQuantities.ItemID
LEFT JOIN TblItem i 
ON OpenOrderItems.ItemID = i.ItemID
WHERE ISNULL(OpenOrderItems.OpenOrderQty,0) > ISNULL(ItemQuantities.TotalQuantityOnHand,0)
GROUP BY OpenOrderItems.ItemID, i.Description
ORDER BY ItemID
```

```sql
CREATE VIEW MostRecentCost1 AS 
(SELECT mrd.ItemID,
ich.LastCost AS LastCost,
MAX(mrd.LastCostDate) AS LastCostDate
FROM MostRecentDate mrd
LEFT JOIN TblItemCostHistory ich
ON mrd.ItemID = ich.ItemID
WHERE ich.LastCostDate = mrd.LastCostDate
GROUP BY mrd.ItemID, ich.LastCost)
```

```sql
With OrderProbability AS (
SELECT
ol.OrderID,
SUM((ol.QtyOrdered*ol.Price) - (ol.QtyOrdered*ISNULL(mrc.LastCost,0))) AS OrderProfit
FROM TblOrderLine ol
LEFT JOIN MostRecentCost1 mrc
ON ol.ItemID = mrc.ItemID
GROUP BY ol.OrderID)
SELECT o.OrderID,
CONVERT(VARCHAR, o.OrderDate, 107) DateOrdered,
c.CustomerID,
c.LastName,
c.FirstName,
CONVERT(DECIMAL(8,2),OrderProbability.OrderProfit) AS OrderProfit
FROM TblOrder o
LEFT JOIN OrderProbability
ON o.OrderID = OrderProbability.OrderID
LEFT JOIN TblCustomer c
ON c.CustomerID = o.CustomerID
```

```sql
CREATE VIEW OrderProfitability AS (
SELECT
ol.OrderID,
SUM((ol.QtyOrdered*ol.Price) - (ol.QtyOrdered*ISNULL(mrc.LastCost,0))) OrderProfit
FROM TblOrderLine ol
LEFT JOIN MostRecentCost1 mrc
ON ol.ItemID = mrc.ItemID
GROUP BY ol.OrderID)
```

```sql
With OrderProbability AS (
SELECT
ol.OrderID,
ol.ItemID,
ol.QtyOrdered,
ol.Price,
ISNULL(mrc.LastCost,0) AS LastCost,
ol.Price - ISNULL(mrc.LastCost,0) AS DIfferenceBetweenPriceAndCost,
ol.QtyOrdered * (ol.Price - ISNULL(mrc.LastCost,0)) AS ExtendedDifference
FROM TblOrderLine ol
LEFT JOIN MostRecentCost1 mrc
ON ol.ItemID = mrc.ItemID
)

SELECT o.OrderID,
CONVERT(VARCHAR, o.OrderDate, 107) DateOrdered,
c.FirstName + ' ' + c.LastName AS CustomerName,
OrderProbability.ItemID,
i.Description ItemDescription,
QtyOrdered,
OrderProbability.Price,
OrderProbability.LastCost,
OrderProbability.DifferenceBetweenPriceAndCost,
OrderProbability.ExtendedDifference
FROM TblOrder o
LEFT JOIN OrderProbability
ON o.OrderID = OrderProbability.OrderID
LEFT JOIN TblCustomer c
ON c.CustomerID = o.CustomerID
LEFT JOIN TblItem i
ON i.ItemID = OrderProbability.ItemID
LEFT JOIN OrderProfitability
ON OrderProfitability.OrderID = OrderProbability.OrderID
WHERE OrderProfitability.OrderProfit = (SELECT MAX(OrderProfit) FROM OrderProfitability)
```

```sql
WITH CaliforniaGameSales AS (
    SELECT
    i.ItemID,
    i.Description,
    SUM(ol.QtyOrdered) AS TotalQtyOrdered,
    AVG(ol.Price) AveragePrice,
    ich.LastCost,
    CONVERT(varchar, ich.LastCostDate, 107) LastCostDate
FROM TblOrderLine ol
LEFT JOIN TblOrder o 
ON o.OrderID = ol.orderid
LEFT JOIN TblCustomer c 
ON c.CustomerID = o.CustomerID
LEFT JOIN TblItem i 
ON i.ItemID = ol.ItemID
LEFT JOIN TblItemCostHistory ich
ON ich.ItemID = i.ItemID
WHERE UPPER(c.CustomerState) = 'CA'
GROUP BY i.ItemID, i.Description, ich.LastCost, ich.LastCostDate
)

SELECT TOP 1
    cgs.ItemID,
    cgs.Description ItemDescription,
    cgs.TotalQtyOrdered,
    AveragePrice,
    LastCost,
    LastCostDate
FROM CaliforniaGameSales cgs
WHERE TotalQtyOrdered = (
    SELECT MAX(TotalQtyOrdered)
    FROM CaliforniaGameSales
)
```
